**Exercícios de Condicionais Aninhadas -  elif**

**Exemplos com elif**

```
# Programa que diz se um numero é par ou impar

num = int(input("Digite um número: "))

if ((num % 2) == 0):
  print("e par")
else:
  print("e impar")
```

```
# Programa que verifica se um número é par e menor que 10 ou se é impar ou se é só par

num = int(input("Digite um número: "))

if ((num % 2) == 0 and num < 10):
  print("e par e menor que 10")
elif (not (num % 2) == 0):
  print("e impar")
else:
  print("e so par")
```

3 - Faça um programa que recebe como entrada a **altura** e a **peso** de uma pessoa, como saída imprima qual a faixa de peso de acordo com o cálculo do **IMC**.

**Informações:**

O cálculo do IMC deve ser feito usando a seguinte fórmula matemática: Peso ÷ altura x altura.

As faixas de acordo com o valor do IMC são:

![Faixas de peso](https://static.tuasaude.com/media/article/wm/sw/imc_32342_l.jpg)


**EXEMPLO DE ENTRADA**


1.65


50

**EXEMPLO DE SAÍDA**

Abaixo do peso



```
altura = float(input("Altura: "))
peso = float(input("Peso: "))

imc = peso / (altura * altura)

if imc <= 18.4 and imc >= 17.0:
  print("Abaixo do peso")
elif imc >= 18.5 and imc <=24.9:
  print("Normal")
elif imc >= 25 and imc <=29.9:
  print("Excesso de peso")
elif imc >= 30:
  print("Obesidade")
  
```

Valor da parcela:  20.0

```
valor = float(input("Digite o valor da compra: "))
forma = input("Digite a forma: ")

if forma == "D":
  print(valor)
elif forma == "P":
  n = int(input("Quantas parcelas?: "))
  valor = valor / n
  print(valor)
```

5 - Faça um programa que verifica se um aluno pode ser monitor ou não de uma matéria. Para que ele seja aprovado ele deve ter tido uma nota final na matéria maior ou igual a 8.0 e deve ter o coeficiente de rendimento maior ou igual a 5.0. Caso ele siga os critérios o programa deve imprimir "APROVADO", caso não, deve imprimir "REPROVADO".

**ENTRADA**

8.5

5.8

**SAÍDA**

APROVADO

----------------------------

**ENTRADA**

7.8
4.9

**SAÍDA**

REPROVADO

```
nota_final = float(input("Digite a nota final: "))
coef = float(input("Digite o coeficiente: "))

if nota_final >= 8.0 and coef >= 5.0:
  print("APROVADO")
elif nota_final < 8.0 and coef < 5.0:
  print("REPROVADO")
```
