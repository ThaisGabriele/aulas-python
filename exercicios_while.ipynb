**Exercícios de while (repetição por condição )**

```
# Exemplo de programa com menu e elif dentro do while
# A palabra reservada break - força sair do while

opcao = 0

while opcao != 7:
  print(''''  [1] Pop
  [2] Rock
  [3] Jazz
  [4] R&B
  [5] Sertanejo
  [6] Digitar um gênero
  [7] Sair
  Qual o seu gênero de música preferido? 
  ''')
  opcao = int(input("Digite o numero da opcao: "))
  
  if opcao == 1:
    print("***Voce escolheu pop**")
  elif opcao == 2:
    print("**Voce escolheu rock**")
  elif opcao == 3:
    print("**Voce escolheu jazz**")
  elif opcao == 4:
    print("**Voce escolheu r&b**")
  elif opcao == 5:
    print("**Voce escolheu sertanejo**")
  elif opcao == 6:
    genero = input("Digite um: ")
    print("Voce digitou {}".format(genero))
  elif opcao == 7:
    break
  else:
    print("Errouu! Opcao invalida, digite novamente")
  
print("Saindo...")
```

1 -  Faça um programa que calcula a soma de cinco números dados como entrada e como saída imprima o a soma desses valores

**ENTRADA**

10

20

30

40

50

**SAIDA**

150

```
soma = qtd = num = 0

while qtd < 5:
  num = int(input("Digite um numero: "))
  soma = soma + num
  qtd += 1
print(soma)
```

```
opcao = 0
num1 = num2 = operacao = 0

while opcao != 5:
  print(''''
        [1] SOMAR
        [2] SUBTRAIR
        [3] DIVIDIR
        [4] MULTIPLICAR
        [5] SUSPENDER
        ''')
        
  opcao = int(input("Digite a operação que você deseja: "))
  num1 = int(input("Digite o num1: "))
  num2 = int(input("Digite o num2: "))
  
  if opcao == 1:
      operacao = num1 + num2
      print(operacao)
  elif opcao == 2:
      operacao = num1 - num2
      print(operacao)
  elif opcao == 3:
      operacao = num1 / num2
      print(operacao)
  elif opcao == 4:
      operacao = num1 * num2
      print(operacao)
  elif opcao == 5:
      break
  else:
     print("Opção inválida. Tenta novamente")
```
