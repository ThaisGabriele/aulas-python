2 - A professora Eulanda tem uma turma com quatro alunos. Após a correção da prova de matemática ela deseja saber qual foi a média da turma.
Faça um programa para calcular a média aritmética da turma de Eulanda e imprima o resultado no formato indicado.

**Exemplo de ENTRADA:**

7.0 

3.6

5.0

10.0

**Exemplo de SAÍDA:**

A média da turma é 6.4

```
nota1 = float(input("Nota 1: "))
nota2 = float(input("Nota 2: "))
nota3 = float(input("Nota 3: "))
nota4 = float(input("Nota 4: "))


media = (nota1 + nota2 + nota3 + nota4) / 4.0

print("A média da turma é ", media)
```
